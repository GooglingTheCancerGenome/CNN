```python
dot_graph = pydot.Dot(graph_type='digraph')

test_bam_node = pydot.Node('Read alignment\n(BAM)')
test_bam_node.set_shape('box3d')
dot_graph.add_node(test_bam_node)

test_gem_node = pydot.Node('GEM mappability track\n(BigWig)')
test_gem_node.set_shape('box3d')
dot_graph.add_node(test_gem_node)

test_2bit_node = pydot.Node('Reference genome\n(2BIT)')
test_2bit_node.set_shape('box3d')
dot_graph.add_node(test_2bit_node)

test_truth_set_vcf_node = pydot.Node('Truth set\n(VCF)')
test_truth_set_vcf_node.set_shape('box3d')
dot_graph.add_node(test_truth_set_vcf_node)

test_truth_set_bedpe_node = pydot.Node('Truth set\n(BEDPE)')
test_truth_set_bedpe_node.set_shape('box3d')
dot_graph.add_node(test_truth_set_bedpe_node)

iedge = pydot.Edge(test_truth_set_vcf_node, test_truth_set_bedpe_node)
iedge.set_label('vcf2bedpe.R')
dot_graph.add_edge(iedge)

clipped_reads_py_node = pydot.Node('clipped_reads.py')
clipped_reads_py_node.set_shape('box')
dot_graph.add_node(clipped_reads_py_node)

clipped_reads_node = pydot.Node('clipped_reads\n(JSON.GZ)')
clipped_reads_node.set_shape('box3d')
dot_graph.add_node(clipped_reads_node)

iedge = pydot.Edge(test_bam_node, clipped_reads_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(clipped_reads_py_node, clipped_reads_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

clipped_read_pos_py_node = pydot.Node('clipped_read_pos.py')
clipped_read_pos_py_node.set_shape('box')
dot_graph.add_node(clipped_read_pos_py_node)

clipped_read_pos_node = pydot.Node('clipped_read_pos\n(JSON.GZ)')
clipped_read_pos_node.set_shape('box3d')
dot_graph.add_node(clipped_read_pos_node)

iedge = pydot.Edge(test_bam_node, clipped_read_pos_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(clipped_read_pos_py_node, clipped_read_pos_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

split_reads_py_node = pydot.Node('split_reads.py')
split_reads_py_node.set_shape('box')
dot_graph.add_node(split_reads_py_node)

split_reads_node = pydot.Node('split_reads\n(JSON.GZ)')
split_reads_node.set_shape('box3d')
dot_graph.add_node(split_reads_node)

iedge = pydot.Edge(test_bam_node, split_reads_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(split_reads_py_node, split_reads_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

clipped_read_distance_py_node = pydot.Node('clipped_read_distance.py')
clipped_read_distance_py_node.set_shape('box')
dot_graph.add_node(clipped_read_distance_py_node)

clipped_read_distance_node = pydot.Node('clipped_read_distance\n(JSON.GZ)')
clipped_read_distance_node.set_shape('box3d')
dot_graph.add_node(clipped_read_distance_node)

iedge = pydot.Edge(test_bam_node, clipped_read_distance_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(clipped_read_distance_py_node, clipped_read_distance_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

coverage_py_node = pydot.Node('coverage.py')
coverage_py_node.set_shape('box')
dot_graph.add_node(coverage_py_node)

coverage_node = pydot.Node('coverage\n(NPY.GZ)')
coverage_node.set_shape('box3d')
dot_graph.add_node(coverage_node)

iedge = pydot.Edge(coverage_py_node, coverage_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

snv_py_node = pydot.Node('snv.py')
snv_py_node.set_shape('box')
dot_graph.add_node(snv_py_node)

snv_node = pydot.Node('snv\n(NPY.GZ)')
snv_node.set_shape('box3d')
dot_graph.add_node(snv_node)

iedge = pydot.Edge(snv_py_node, snv_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

chr_array_py_node = pydot.Node('chr_array.py')
chr_array_py_node.set_shape('box')
dot_graph.add_node(chr_array_py_node)

chr_array_node = pydot.Node('chr_array\n(BCOLZ)')
chr_array_node.set_shape('box3d')
dot_graph.add_node(chr_array_node)

iedge = pydot.Edge(clipped_reads_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(clipped_read_pos_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(split_reads_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(clipped_read_distance_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(coverage_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(snv_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(test_gem_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(test_2bit_node, chr_array_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(chr_array_py_node, chr_array_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

label_sr_py_node = pydot.Node('label_windows.py')
label_sr_py_node.set_shape('box')
dot_graph.add_node(label_sr_py_node)

iedge = pydot.Edge(test_bam_node, label_sr_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(split_reads_node, label_sr_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(test_truth_set_bedpe_node, label_sr_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

label_sr_node = pydot.Node('labels\n(JSON.GZ)')
label_sr_node.set_shape('box3d')
dot_graph.add_node(label_sr_node)

iedge = pydot.Edge(label_sr_py_node, label_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

create_window_pairs_py_sr_node = pydot.Node('create_window_pairs.py')
create_window_pairs_py_sr_node.set_shape('box')
dot_graph.add_node(create_window_pairs_py_sr_node)

add_win_channels = pydot.Node('add_win_channels.py')
add_win_channels.set_shape('box')
dot_graph.add_node(add_win_channels)

train_py_sr_node = pydot.Node('train.py\n')
train_py_sr_node.set_shape('box')
dot_graph.add_node(train_py_sr_node)

predict_py_sr_node = pydot.Node('predict.py\n')
predict_py_sr_node.set_shape('box')
dot_graph.add_node(predict_py_sr_node)

iedge = pydot.Edge(test_bam_node, label_sr_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(test_truth_set_bedpe_node, label_sr_py_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(label_sr_node, create_window_pairs_py_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(chr_array_node, create_window_pairs_py_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

windows_sr_node = pydot.Node('windows\n(NPZ)')
windows_sr_node.set_shape('box3d')
dot_graph.add_node(windows_sr_node)

iedge = pydot.Edge(create_window_pairs_py_sr_node, windows_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

windows_en_node = pydot.Node('windows_en\n(NPZ)')
windows_en_node.set_shape('box3d')
dot_graph.add_node(windows_en_node)

iedge = pydot.Edge(test_bam_node, add_win_channels)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(windows_sr_node, add_win_channels)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(add_win_channels, windows_en_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

windows_train_node = pydot.Node('Training set windows_en\n(NPZ)')
windows_train_node.set_shape('box3d')
dot_graph.add_node(windows_train_node)

windows_test_node = pydot.Node('Test set windows_en\n(NPZ)')
windows_test_node.set_shape('box3d')
dot_graph.add_node(windows_test_node)

model = pydot.Node('Model\n(HDF5)')
model.set_shape('box3d')
dot_graph.add_node(model)

iedge = pydot.Edge(windows_train_node, train_py_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(train_py_sr_node, model)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(windows_test_node, predict_py_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(model, predict_py_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

predictions_sr_node = pydot.Node('SV calls\n(BEDPE)')
predictions_sr_node.set_shape('box3d')
dot_graph.add_node(predictions_sr_node)

merge_sv_calls = pydot.Node('merge_sv_calls.R')
merge_sv_calls.set_shape('box')
dot_graph.add_node(merge_sv_calls)

bedpe_to_vcf = pydot.Node('bedpe_to_vcf.py')
bedpe_to_vcf.set_shape('box')
dot_graph.add_node(bedpe_to_vcf)

predictions_vcf_node = pydot.Node('SV calls\n(VCF)')
predictions_vcf_node.set_shape('box3d')
dot_graph.add_node(predictions_vcf_node)

iedge = pydot.Edge(predict_py_sr_node, predictions_sr_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(predictions_sr_node, merge_sv_calls)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(merge_sv_calls, bedpe_to_vcf)
iedge.set_label('')
dot_graph.add_edge(iedge)

iedge = pydot.Edge(bedpe_to_vcf, predictions_vcf_node)
iedge.set_label('')
dot_graph.add_edge(iedge)

dot_graph.write_svg('cnn.svg')
# dot_graph.write_ps2('cnn.ps2')
SVG('cnn.svg')
```
